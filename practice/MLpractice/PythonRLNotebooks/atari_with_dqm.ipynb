```python
from stable_baselines3 import DQN 
from stable_baselines3.common.env_util import make_atari_env
from stable_baselines3.common.vec_env import VecFrameStack
import ale_py
```

```python
env = make_atari_env("ALE/Pacman-v5", n_envs=1, seed=0)
```

```python
model = DQN(policy="CnnPolicy", env=env, verbose=1, learning_rate=1e-4, buffer_size=100000, learning_starts=100, 
            batch_size=32, gamma=.99, exploration_fraction=0.1)
model.learn(total_timesteps=2_000_000)
```

```python
model.save("dqn_pacman")
```

```python
from stable_baselines3.common.evaluation import evaluate_policy
from stable_baselines3.common.vec_env import VecVideoRecorder, DummyVecEnv
```

```python
# Load the trained agent
model = DQN.load("dqn_pacman", env=env)

# Evaluate the agent
mean_reward, std_reward = evaluate_policy(model, model.get_env(), n_eval_episodes=5)
print(f"Mean reward: {mean_reward} +/- {std_reward}")
```

```python
vec_env = model.get_env()
obs = vec_env.reset()
for i in range(10000):
    action, _states = model.predict(obs, deterministic=True)
    obs, rewards, dones, info = vec_env.step(action)
    vec_env.render("human")
    if i % 1000 == 0:
        print(f"Step {i}: Action taken: {action}, Reward: {rewards}, Done: {dones}")
```
